```python
#1
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

########################################
# Generate meaningful data
########################################
np.random.seed(0)
x = np.random.uniform(50, 80, 25)
noise = np.random.normal(0, np.sqrt(20), 25)
t = 3.86 * x - 110.42 + noise

########################################
# Plot scatter plot of data
########################################
plt.scatter(x, t)
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.title('Scatter plot of Height vs Weight')
plt.show()

########################################
# Weight estimation through error surface
########################################
def Error(w, t, x):
    # Extracting w0 and w1 from the weight vector w
    w0, w1 = w[0], w[1]

    # Calculate the estimated target values
    estimated_t = w0 + w1 * x

    # Calculate the mean squared error
    error = np.mean((estimated_t - t) ** 2)

    return error

# Sample a bunch of w's around w_opt and compute the associated error
w0_values = np.linspace(-500, 500, 100)
w1_values = np.linspace(-10, 10, 100)

W0, W1 = np.meshgrid(w0_values, w1_values)
error_surface = np.zeros_like(W0)
for i in range(len(w1_values)):
    for j in range(len(w0_values)):
        error_surface[i, j] = Error([w0_values[j], w1_values[i]], t, x)

# Plot 3D error surface
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W0, W1, error_surface, cmap='viridis')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Error')
ax.set_title('Error Surface')
plt.show()

# Locate the minima of the error surface
min_error_index = np.unravel_index(np.argmin(error_surface), error_surface.shape)
min_w0 = w0_values[min_error_index[1]]
min_w1 = w1_values[min_error_index[0]]
print("Minimum error occurs at (w0, w1) =", (min_w0, min_w1))

########################################
# Least squares approach to estimate the weights
########################################
def LinearRegression(x, t):
    X = np.vstack((np.ones_like(x), x)).T
    w_opt = np.linalg.inv(X.T @ X) @ X.T @ t
    return w_opt

# Estimate optimal weights using "LinearRegression" function
w_opt = LinearRegression(x, t)

# Estimate the targets using the input x and the estimated weights
estimated_targets = w_opt[0] + w_opt[1] * x

########################################
# Plot the estimated line on top of the above scatter plot
########################################
plt.scatter(x, t, label='Data')
plt.plot(x, estimated_targets, color='red', label='Estimated Line')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.title('Linear Regression')
plt.legend()
plt.show()

########################################
# Compare the estimated weights using least squares approach with the error surface approach
########################################
print("Optimal weights using least squares approach:", w_opt)
print("Optimal weights using error surface approach:", (min_w0, min_w1))

# Compare errors from both methods
error_least_squares = Error(w_opt, t, x)
print("Error using least squares approach:", error_least_squares)
print("Error using error surface approach:", error_surface[min_error_index])
```

<b> Report your observations </b>

Observations:
Error Surface vs. Least Squares:

Both the error surface and least squares approaches found optimal weights for the linear regression model that fits the relationship between height and weight.
However, the optimal weights differ slightly:
Error surface: (-116.1616, 3.9394)
Least squares: (-102.8851, 3.7576)
This difference suggests that the error surface method is slightly more sensitive to the data and found a solution that minimizes the error slightly better than the least squares method, which finds an average solution based on all data points.

It would be beneficial to quantitatively compare the errors associated with both sets of weights to better understand the practical significance of the difference. You could calculate the mean squared error for each set of weights and see which one results in a lower error on the data.

```python
#2
########################################
# Generate 20 data points
########################################
np.random.seed(0)
x = np.random.rand(20)
e = np.random.normal(0, 0.1, 20)
t = np.sin(2 * np.pi * x) + e

# Obtain train and test splits
x_train, x_test = x[:10], x[10:]
t_train, t_test = t[:10], t[10:]

########################################
# Fitting Mth degree polynomial using least squares approach
########################################
def PolynomialFit(X_train, Y_train, M, lamda=0):
    X = np.vander(X_train, M + 1, increasing=True)
    if lamda == 0:
        w_opt = np.linalg.inv(X.T @ X) @ X.T @ Y_train
    else:
        w_opt = np.linalg.inv(X.T @ X + lamda * np.identity(M + 1)) @ X.T @ Y_train
    return w_opt

def PolynomialPred(w_est, X):
    X = np.vander(X, len(w_est), increasing=True)
    Y_pred = X @ w_est
    return Y_pred

def PolynomialPred_Error(w_est, X_train, X_test, Y_train, Y_test):
    Y_train_pred = PolynomialPred(w_est, X_train)
    Y_test_pred = PolynomialPred(w_est, X_test)
    TrainError = np.sqrt(np.mean((Y_train_pred - Y_train) ** 2))
    TestError = np.sqrt(np.mean((Y_test_pred - Y_test) ** 2))
    return TrainError, TestError

# Iterate through range of M values
M_range = list(range(10))
TrError, TeError = [], []
for M in M_range:
    w_opt = PolynomialFit(x_train, t_train, M)
    TrainError, TestError = PolynomialPred_Error(w_opt, x_train, x_test, t_train, t_test)
    TrError.append(TrainError)
    TeError.append(TestError)


# Plot training and testing estimates along with original targets
plt.scatter(x_train, t_train, label='Training Data', color='blue')
plt.scatter(x_test, t_test, label='Testing Data', color='red')
x_values = np.linspace(0, 1, 100)
for M in M_range:
    w_opt = PolynomialFit(x_train, t_train, M)
    y_values = PolynomialPred(w_opt, x_values)
    plt.plot(x_values, y_values, label=f'M={M}')
plt.xlabel('x')
plt.ylabel('t')
plt.title('Polynomial Fits with Different Model Orders')
plt.legend()
plt.show()

# Plot training error versus polynomial order and testing error versus polynomial order
plt.plot(M_range, TrError, label='Training Error')
plt.plot(M_range, TeError, label='Testing Error')
plt.xlabel('Model Order (M)')
plt.ylabel('Root Mean Squared Error')
plt.title('Training and Testing Error vs Polynomial Order')
plt.legend()
plt.show()

########################################
# Increase the size of training data set to 100 points and repeat the experiments
########################################
np.random.seed(0)
x_large = np.random.rand(100)
e_large = np.random.normal(0, 0.1, 100)
t_large = np.sin(2 * np.pi * x_large) + e_large
x_train_large, x_test_large = x_large[:50], x_large[50:]
t_train_large, t_test_large = t_large[:50], t_large[50:]

M_range_large = list(range(10))
TrError_large, TeError_large = [], []
for M in M_range_large:
    w_opt_large = PolynomialFit(x_train_large, t_train_large, M)
    TrainError_large, TestError_large = PolynomialPred_Error(w_opt_large, x_train_large, x_test_large, t_train_large, t_test_large)
    TrError_large.append(TrainError_large)
    TeError_large.append(TestError_large)

plt.plot(M_range_large, TrError_large, label='Training Error')
plt.plot(M_range_large, TeError_large, label='Testing Error')
plt.xlabel('Model Order (M)')
plt.ylabel('Root Mean Squared Error')
plt.title('Effect of Model Order on Training and Testing Error (Large Dataset)')
plt.legend()
plt.show()

########################################
# Effect of regularization
########################################
lamdas = [0, 1e-7, 1e-4, 1e-2, 1]
TrError_reg, TeError_reg = [], []

for lamda in lamdas:
    TrError_lamda, TeError_lamda = [], []
    for M in M_range_large:
        w_opt_reg = PolynomialFit(x_train_large, t_train_large, M, lamda)
        TrainError_reg, TestError_reg = PolynomialPred_Error(w_opt_reg, x_train_large, x_test_large, t_train_large, t_test_large)
        TrError_lamda.append(TrainError_reg)
        TeError_lamda.append(TestError_reg)
    TrError_reg.append(TrError_lamda)
    TeError_reg.append(TeError_lamda)

for i, lamda in enumerate(lamdas):
    plt.plot(M_range_large, TrError_reg[i], label=f'Training Error (lambda={lamda})')
    plt.plot(M_range_large, TeError_reg[i], label=f'Testing Error (lambda={lamda})')

plt.xlabel('Model Order (M)')
plt.ylabel('Root Mean Squared Error')
plt.title('Effect of Model Order and Regularization on Training and Testing Error')
plt.legend()
plt.show()

########################################
# Effect of bias regularization
########################################
def PolynomialFitWithBias_Ridge(X_train, Y_train, M, lamda=0):
    X = np.vander(X_train, M + 1, increasing=True)
    X = np.column_stack((X, np.ones_like(X[:, 0])))  # Adding bias column
    I = np.identity(M + 2)
    I[-1, -1] = 0  # Do not penalize bias term
    try:
        w_opt = np.linalg.inv(X.T @ X + lamda * I) @ X.T @ Y_train
    except np.linalg.LinAlgError:
        print(f"Singular matrix encountered for M={M} and lambda={lamda}.")
        return None
    return w_opt

def PolynomialPred(w_est, X):
    X = np.vander(X, len(w_est) - 1, increasing=True)
    X = np.column_stack((X, np.ones_like(X[:, 0])))  # Adding bias column
    Y_pred = X @ w_est
    return Y_pred

# Attempt polynomial fitting for different model complexities and regularization parameters
M_values = [5, 7, 9]
lambda_values = [0, 1e-7, 1e-4, 1e-2, 1]
w_opts = {}

for M in M_values:
    for lamda in lambda_values:
        w_opt = PolynomialFitWithBias_Ridge(x_train, t_train, M, lamda)
        if w_opt is not None:
            w_opts[(M, lamda)] = w_opt

# Plotting
plt.figure(figsize=(12, 8))

for (M, lamda), w_opt in w_opts.items():
    x_values = np.linspace(0, 1, 100)
    y_values = PolynomialPred(w_opt, x_values)
    label = f'M={M}, lambda={lamda}'
    plt.plot(x_values, y_values, label=label)

plt.scatter(x_train, t_train, label='Training Data', color='blue')
plt.scatter(x_test, t_test, label='Testing Data', color='red')
plt.xlabel('x')
plt.ylabel('t')
plt.title('Estimated Polynomials with Different Model Complexity and Regularization')
plt.legend()
plt.show()
```

<b> Report your observations </b>

Key Points:

Overfitting: Lines without regularization (lambda=0) show initial error reduction with increasing model complexity (M), but eventually rise again due to overfitting.
Regularization Effect: Increased regularization (higher lambda) flattens the lines, reducing sensitivity to M and mitigating overfitting.
Singular Matrices: Some combinations of M and lambda cause missing lines due to encountering singular matrices (numerical instability).
Overall:

Regularization helps prevent overfitting: Use of regularization (lambda > 0) reduces overfitting compared to no regularization.
Choose lambda carefully: Optimal lambda depends on the dataset and model complexity.
Avoid singular matrices: Certain combinations of M and lambda might lead to unstable results.



```python
#3
# Generate 100 data points
np.random.seed(0)
X = np.random.rand(100)
e = np.random.normal(0, 0.1, 100)

# Define new target function
def new_target_function(X):
    t = np.zeros_like(X)
    t[(X >= 0) & (X < 1)] = np.sin(2 * np.pi * X[(X >= 0) & (X < 1)])
    t[(X >= 1) & (X < 2)] = np.abs(X[(X >= 1) & (X < 2)] - 1)
    t[(X >= 2) & (X < 3)] = np.exp(-(X[(X >= 2) & (X < 3)] - 2)**2)
    return t

t = new_target_function(X)

# Split data into training and testing sets
X_train = X[::2]
t_train = t[::2]
X_test = X[1::2]
t_test = t[1::2]

# Function to estimate the parameters
def KernelRegressionFit(X_train, Y_train, kernelType, M, lamda):
    if kernelType == 'polynomial':
        X = np.vander(X_train, M+1, increasing=True)
    elif kernelType == 'gaussian':
        # Using Gaussian kernel
        X = np.exp(-(X_train[:, None] - X_train[None, :])**2 / (2 * lamda**2))
    elif kernelType == 'sigmoidal':
        # Using Sigmoid kernel
        X = np.tanh(lamda * (X_train[:, None] @ X_train[None, :]) + 1)
    else:
        raise ValueError("Invalid kernel type")

    # Estimate weights using Pseudo inverse solution
    w_opt = np.linalg.pinv(X) @ Y_train

    return w_opt

# Function to compute the training and testing errors
def KernelRegressionPred_Error(w_est, X_train, Y_train, X_test, Y_test, kernelType, lamda):
    if kernelType == 'polynomial':
        X_train_k = np.vander(X_train, len(w_est), increasing=True)
        X_test_k = np.vander(X_test, len(w_est), increasing=True)
    elif kernelType == 'gaussian':
        # Using Gaussian kernel
        X_train_k = np.exp(-(X_train[:, None] - X_train[None, :])**2 / (2 * lamda**2))
        X_test_k = np.exp(-(X_test[:, None] - X_train[None, :])**2 / (2 * lamda**2))
    elif kernelType == 'sigmoidal':
        # Using Sigmoid kernel
        X_train_k = np.tanh(lamda * (X_train[:, None] @ X_train[None, :]) + 1)
        X_test_k = np.tanh(lamda * (X_test[:, None] @ X_train[None, :]) + 1)
    else:
        raise ValueError("Invalid kernel type")

    # Estimate training and testing targets
    Y_train_pred = X_train_k @ w_est
    Y_test_pred = X_test_k @ w_est

    # Compute training and testing errors
    TrainError = np.mean((Y_train_pred - Y_train)**2)
    TestError = np.mean((Y_test_pred - Y_test)**2)

    return TrainError, TestError

# Iterate through range of M values
M_range = list(range(10))

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoid_tr_error = []
sigmoid_te_error = []

lamda = 0.1

for M in M_range:
    # Fit Mth order polynomial using three kernels i.e {Polynomial, Gaussian, Sigmoid}
    w_polynomial = KernelRegressionFit(X_train, t_train, 'polynomial', M, lamda)
    w_gaussian = KernelRegressionFit(X_train, t_train, 'gaussian', M, lamda)
    w_sigmoid = KernelRegressionFit(X_train, t_train, 'sigmoidal', M, lamda)

    # Predict training and testing targets using estimated w
    poly_tr_err, poly_te_err = KernelRegressionPred_Error(w_polynomial, X_train, t_train, X_test, t_test, 'polynomial', lamda)
    gaussian_tr_err, gaussian_te_err = KernelRegressionPred_Error(w_gaussian, X_train, t_train, X_test, t_test, 'gaussian', lamda)
    sigmoid_tr_err, sigmoid_te_err = KernelRegressionPred_Error(w_sigmoid, X_train, t_train, X_test, t_test, 'sigmoidal', lamda)

    # Store errors for plotting
    polynomial_tr_error.append(poly_tr_err)
    polynomial_te_error.append(poly_te_err)
    gaussian_tr_error.append(gaussian_tr_err)
    gaussian_te_error.append(gaussian_te_err)
    sigmoid_tr_error.append(sigmoid_tr_err)
    sigmoid_te_error.append(sigmoid_te_err)

# Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_train, t_train, color='blue', label='Training Data')
plt.scatter(X_test, t_test, color='red', label='Testing Data')
plt.title('Original Data')
plt.legend()

# Plot training error vs polynomial order and testing error vs polynomial order for all three different kernels
plt.subplot(1, 2, 2)
plt.plot(M_range, polynomial_tr_error, marker='o', color='blue', label='Polynomial Training Error')
plt.plot(M_range, polynomial_te_error, marker='o', color='cyan', label='Polynomial Testing Error')
plt.plot(M_range, gaussian_tr_error, marker='x', color='green', label='Gaussian Training Error')
plt.plot(M_range, gaussian_te_error, marker='x', color='lime', label='Gaussian Testing Error')
plt.plot(M_range, sigmoid_tr_error, marker='^', color='purple', label='Sigmoidal Training Error')
plt.plot(M_range, sigmoid_te_error, marker='^', color='magenta', label='Sigmoidal Testing Error')
plt.title('Training and Testing Errors')
plt.xlabel('Model Order (M)')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.show()
```

Kernel Impact:

Gaussian: Achieves balanced training and testing error, indicating good generalization. Shows less overfitting than other kernels.

Polynomial: Prone to significant overfitting, especially at higher model orders. Training error decreases quickly but testing error stays high, suggesting poor generalization.

Sigmoid: Exhibits consistently high testing errors across all model orders, suggesting difficulty capturing data complexity.
Model Order Impact:

For all kernels, increasing model order generally reduces training error, but also increases risk of overfitting, reflected in rising training errors for polynomial and sigmoid kernels at higher orders.
Optimal model order depends on data and kernel, where a balance between training and testing error is desired. Gaussian kernel seems to achieve this balance around M=6-8.

```python
#4
# Define the error prediction function
def ErrorPred(w_est, X_train, Y_train, X_test, Y_test, kernelType):
    # Initialize variables to store kernel representations
    X_tr = None
    X_te = None

    # Calculate the degree of the polynomial
    M = len(w_est) - 1

    # Calculate kernel representations based on the kernel type
    if kernelType == 'polynomial':
        # Use polynomial kernel to transform the data
        X_tr = np.vander(X_train, M+1, increasing=True)
        X_te = np.vander(X_test, M+1, increasing=True)
    elif kernelType == 'gaussian':
        # Use Gaussian kernel to transform the data
        X_tr = np.exp(-(X_train[:, None] - X_train[None, :])**2 / (2 * lamda**2))
        X_te = np.exp(-(X_test[:, None] - X_train[None, :])**2 / (2 * lamda**2))
    elif kernelType == 'sigmoidal':
        # Use sigmoidal kernel to transform the data
        X_tr = np.tanh(lamda * (X_train[:, None] @ X_train[None, :]) + 1)
        X_te = np.tanh(lamda * (X_test[:, None] @ X_train[None, :]) + 1)
    else:
        raise ValueError("Invalid kernel type")

    # Compute predicted targets for training and testing data
    Y_train_pred = X_tr @ w_est
    Y_test_pred = X_te @ w_est

    # Compute training and testing errors
    TrainError = np.mean((Y_train_pred - Y_train)**2)
    TestError = np.mean((Y_test_pred - Y_test)**2)

    return TrainError, TestError

# Define the online training function
def OnlineTraining(X_train, Y_train, X_test, Y_test, kernelType, M, Epochs, BatchSize, stepSize):
    # Initialize the weights
    weights = np.zeros(M + 1)

    # Initialize lists to store errors for plotting
    train_errors = []
    test_errors = []

    # Iterate through epochs
    for epoch in range(Epochs):
        # Compute the train and test errors using the current weights
        tr_err, te_err = ErrorPred(weights, X_train, Y_train, X_test, Y_test, kernelType)
        train_errors.append(tr_err)
        test_errors.append(te_err)

        # Shuffle the data
        permutation = np.random.permutation(len(X_train))
        X_train_shuffled = X_train[permutation]
        Y_train_shuffled = Y_train[permutation]

        # Iterate through the batches
        for i in range(0, len(X_train_shuffled), BatchSize):
            X_batch = X_train_shuffled[i:i+BatchSize]
            Y_batch = Y_train_shuffled[i:i+BatchSize]

            # Compute kernel representation
            if kernelType == 'polynomial':
                X_batch_k = np.vander(X_batch, M+1, increasing=True)
            elif kernelType == 'gaussian':
                X_batch_k = np.exp(-(X_batch[:, None] - X_train_shuffled[None, :])**2 / (2 * lamda**2))
            elif kernelType == 'sigmoidal':
                X_batch_k = np.tanh(lamda * (X_batch[:, None] @ X_train_shuffled[None, :]) + 1)

            # Compute the gradient of weights
            errors = X_batch_k @ weights - Y_batch
            gradients = np.mean(X_batch_k.T * errors, axis=1)

            # Update the weights using SGD
            weights -= stepSize * gradients

    # Plot training and testing error across the epochs
    plt.plot(range(1, Epochs + 1), train_errors, label='Training MSE')
    plt.plot(range(1, Epochs + 1), test_errors, label='Testing MSE')
    plt.xlabel('Epochs')
    plt.ylabel('Mean Squared Error')
    plt.title('Training and Testing Error vs. Epochs')
    plt.legend()
    plt.show()

    # Return the estimated weights and errors
    return weights, train_errors, test_errors

# Define the synthetic data generation function
def generate_data():
    np.random.seed(0)
    X_train = np.random.rand(100)
    Y_train = np.sin(2 * np.pi * X_train) + np.random.normal(0, 0.1, 100)
    X_test = np.random.rand(50)
    Y_test = np.sin(2 * np.pi * X_test) + np.random.normal(0, 0.1, 50)
    return X_train, Y_train, X_test, Y_test

# Parameters
kernelType = 'polynomial'  # or 'gaussian' or 'sigmoidal'
M = 9  # Polynomial degree or number of Gaussian/Sigmoidal basis functions
Epochs = 1000
lamda = 0.1  # For Gaussian and Sigmoidal kernels
BatchSize = 1000
stepSize = 0.001

# Repeat 3(a) with online training
X_train, Y_train, X_test, Y_test = generate_data()
estimated_weights_a,train_errors_a, test_errors_a = OnlineTraining(X_train, Y_train, X_test, Y_test, kernelType, M, Epochs, BatchSize, stepSize)

# Repeat 3(b) with online training
estimated_weights_b,train_errors_b, test_errors_b = OnlineTraining(X_train, Y_train, X_test, Y_test, kernelType, M, Epochs, BatchSize, stepSize)

# Plot training and testing error across the epochs for 3(a) with online training
plt.plot(train_errors_a, label='Training Error (3a)')
plt.plot(test_errors_a, label='Testing Error (3a)')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Online Training: Error Across Epochs (3a)')
plt.legend()
plt.show()

# Plot training and testing error across the epochs for 3(b) with online training
plt.plot(train_errors_b, label='Training Error (3b)')
plt.plot(test_errors_b, label='Testing Error (3b)')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Online Training: Error Across Epochs (3b)')
plt.legend()
plt.show()
```

Observations:

General Behavior:

Both plots show training error generally decreasing with epochs, suggesting the model learns the training data.
Testing error in Plot 1 increases later, indicating overfitting, while it remains constant in Plot 2, suggesting underfitting.


Step Size:

Smaller values: slower but stable convergence, less likely to overshoot.
Larger values: faster but potentially unstable, overshoots or oscillates.

Batch Size:

Larger values: faster convergence, stable gradient estimates.
Smaller values: slower but potentially avoids local minima, noisy gradient estimates.

```python
#5
# Function to generate noisy sinusoidal data
def generate_data(N):
    X = np.linspace(0, 2*np.pi, N)
    y = np.sin(X) + np.random.normal(0, 0.3, N)  # Adding Gaussian noise
    return X, y

# Function to compute Gaussian basis functions
def gaussian_basis(x, mu, sigma=0.1):
    return np.exp(-0.5 * ((x - mu) / sigma)**2)

# Function to fit regularized least squares regression
def fit_regression(X, y, M, lambd):
    Phi = np.array([gaussian_basis(X, mu) for mu in np.linspace(0, 2*np.pi, M)])
    w = np.linalg.inv(Phi @ Phi.T + lambd * np.eye(M)) @ Phi @ y
    return w

# Function to evaluate the model on test data
def evaluate_model(X_test, w):
    Phi_test = np.array([gaussian_basis(X_test, mu) for mu in np.linspace(0, 2*np.pi, len(w))])
    y_pred = Phi_test.T @ w
    return y_pred

# Generate 100 datasets and fit models
L = 100  # Number of datasets
N = 25   # Number of data points per dataset
M = 25   # Order of the linear regression model
lambdas = [0.01, 0.1, 1.0]  # Low, middle, and high regularization coefficients

plt.figure(figsize=(15, 10))

for i, lambd in enumerate(lambdas):
    plt.subplot(3, 2, 2*i+1)
    plt.title(f"100 Estimated Curves (λ={lambd})")
    plt.xlabel("X")
    plt.ylabel("Y")

    plt.subplot(3, 2, 2*i+2)
    plt.title(f"Mean Curve (λ={lambd})")
    plt.xlabel("X")
    plt.ylabel("Y")

    mean_curve = np.zeros(N)

    for j in range(L):
        X, y = generate_data(N)
        w = fit_regression(X, y, M, lambd)
        y_pred = evaluate_model(X, w)

        plt.subplot(3, 2, 2*i+1)
        plt.plot(X, y_pred, color='gray', alpha=0.3)

        mean_curve += y_pred

    mean_curve /= L
    plt.subplot(3, 2, 2*i+2)
    plt.plot(X, mean_curve, color='blue', label='Mean Curve')
    plt.plot(X, np.sin(X), color='red', linestyle='--', label='Original Function')
    plt.legend()

plt.tight_layout()
plt.show()
```

<b> Report your observations </b>

As the regularization parameter (λ) increases, the variance of the estimated curves decreases.

As the regularization parameter (λ) increases, the bias of the estimated curves increases, meaning they tend to deviate more from the true underlying function.

There is a trade-off between variance and bias in regularized least squares regression. Choosing a higher λ reduces variance but increases bias, and vice versa.

The mean curve obtained by averaging the estimated curves is closer to the original function for lower values of λ. This is because higher λ values lead to higher bias, which can cause the estimated curves to deviate from the true function in different directions.



```python
#6
# Function to generate noisy sinusoidal data
def generate_data(N):
    X = np.linspace(0, 2*np.pi, N)
    y = np.sin(X) + np.random.normal(0, 0.3, N)  # Adding Gaussian noise
    return X, y

# Function to compute Gaussian basis functions
def gaussian_basis(x, mu, sigma=0.1):
    return np.exp(-0.5 * ((x - mu) / sigma)**2)

# Function to update posterior statistics
def update_posterior(X, y, M, prior_mean, prior_covariance):
    Phi = np.array([gaussian_basis(X, mu) for mu in np.linspace(0, 2*np.pi, M)])
    S_N_inv = prior_covariance
    m_N = prior_mean

    for i in range(len(X)):
        phi_i = Phi[:, i].reshape(-1, 1)
        S_N_inv += np.outer(phi_i, phi_i)
        m_N += y[i] * phi_i

    S_N = np.linalg.inv(S_N_inv)
    return m_N, S_N

# Function to sample weight vector from posterior distribution
def sample_posterior(m_N, S_N):
    return np.random.multivariate_normal(m_N.ravel(), S_N).reshape(-1, 1)

# Function to evaluate the curve fit
def evaluate_curve(X, w_sampled):
    Phi = np.array([gaussian_basis(X, mu) for mu in np.linspace(0, 2*np.pi, len(w_sampled))])
    return Phi.T @ w_sampled


# Generate noisy sinusoidal data
N = 100
X, y = generate_data(N)

# Prior parameters for standard normal prior
prior_mean = np.zeros((20, 1))
prior_covariance = np.eye(20)

########################################
# Update the statistics of posterior density
########################################

# Iterate through the data points and update the stats of posterior density
m_N, S_N = update_posterior(X, y, 20, prior_mean, prior_covariance)
for i in range(N):
    m_N, S_N = update_posterior(X[:i+1], y[:i+1], 20, m_N, S_N)

########################################
# Sample weight vector from posterior distribution. Estimate the curve, repeat the procedure for 100 times and get the avg fit
########################################

# Sample weight vector from posterior distribution, estimate the curve, repeat the procedure for 100 times and get the avg fit
num_samples = 100
curve_fits = []
for _ in range(num_samples):
    w_sampled = sample_posterior(m_N, S_N)
    curve_fit = evaluate_curve(X, w_sampled)
    curve_fits.append(curve_fit)

########################################
# Predictive distribution analysis
########################################

# Predictive distribution analysis through sampling
plt.figure(figsize=(10, 6))
for i in range(1, 26):
    plt.subplot(5, 5, i)
    plt.title(f'N = {i}')
    plt.xlabel('X')
    plt.ylabel('Y')
    for _ in range(10):
        w_sampled = sample_posterior(m_N, S_N)
        curve_fit = evaluate_curve(X[:i], w_sampled[:i])
        plt.plot(X[:i], curve_fit, color='gray', alpha=0.3)
    plt.scatter(X[:i], y[:i], color='blue')
plt.tight_layout()
plt.show()

# Predictive distribution analysis through variance
plt.figure(figsize=(10, 6))
for i in range(1, 26):
    plt.subplot(5, 5, i)
    plt.title(f'N = {i}')
    plt.xlabel('X')
    plt.ylabel('Y')
    mean_curve_fit = np.mean(curve_fits, axis=0)
    std_curve_fit = np.std(curve_fits, axis=0)
    plt.plot(X[:i], mean_curve_fit[:i], color='green', label='Mean Curve Fit')
    plt.fill_between(X[:i], mean_curve_fit[:i].ravel() - 2 * std_curve_fit[:i].ravel(), mean_curve_fit[:i].ravel() + 2 * std_curve_fit[:i].ravel(), color='orange', alpha=0.3, label='±2σ')
    plt.scatter(X[:i], y[:i], color='blue')
    plt.legend()
plt.tight_layout()
plt.show()
```

-The curve shows a narrowing 2σ confidence interval around the mean curve fit with increasing data size. This confirms the reduced uncertainty and reinforces the trend observed in the first plot.
